```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_circles, make_moons
```

```python
def plot_decision_boundary(model, X, y, title):
    plt.figure(figsize=(6,5))

    # create grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # plot boundary
    plt.contourf(xx, yy, Z, alpha=0.3)

    # plot data
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')

    plt.title(title)
    plt.show()
```

Data Is circular

```python
# -----------------------------
# STEP 1: Create Non-Linear Data
# -----------------------------
X, y = make_circles(n_samples=200, noise=0.1, factor=0.5)


# -----------------------------
# STEP 2: Train SVM with different kernels
# -----------------------------

# 1. Linear Kernel
model_linear = svm.SVC(kernel='linear')
model_linear.fit(X, y)
plot_decision_boundary(model_linear, X, y, "Linear Kernel")

# 2. Polynomial Kernel
model_poly = svm.SVC(kernel='poly', degree=3)
model_poly.fit(X, y)
plot_decision_boundary(model_poly, X, y, "Polynomial Kernel (degree=3)")

# 3. RBF Kernel
model_rbf = svm.SVC(kernel='rbf', gamma=10)
model_rbf.fit(X, y)
plot_decision_boundary(model_rbf, X, y, "RBF Kernel")

# 4. Sigmoid Kernel
model_sigmoid = svm.SVC(kernel='sigmoid')
model_sigmoid.fit(X, y)
plot_decision_boundary(model_sigmoid, X, y, "Sigmoid Kernel")
```

```python
# 3. RBF Kernel
model_rbf = svm.SVC(kernel='rbf', gamma=1)
model_rbf.fit(X, y)
plot_decision_boundary(model_rbf, X, y, "RBF Kernel")
```

```python
# 3. RBF Kernel
model_rbf = svm.SVC(kernel='rbf',C=0.001)
model_rbf.fit(X, y)
plot_decision_boundary(model_rbf, X, y, "RBF Kernel")
```

```python
# 3. RBF Kernel
model_rbf = svm.SVC(kernel='rbf',C=1000)
model_rbf.fit(X, y)
plot_decision_boundary(model_rbf, X, y, "RBF Kernel")
```
