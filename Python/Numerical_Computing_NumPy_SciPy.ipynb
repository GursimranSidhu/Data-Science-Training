# Part 2 — SciPy Scientific Computing

## Numerical Integration

```python
f = lambda x: x**2
result, error = integrate.quad(f, 0, 1)
print('Integral result:', result)
```

## Solving Differential Equations

```python
def model(t,y):    
    return -2*y
sol = solve_ivp(model,[0,5],[1], t_eval=np.linspace(0,5,100))
plt.plot(sol.t, sol.y[0])
plt.title('dy/dt = -2y')
plt.show()
```

## Optimization

```python
f = lambda x: x**2 + 5*x + 6
result = optimize.minimize(f, 0)
print('Minimum at:', result.x)
```

## Interpolation

```python
x = np.array([0,1,2,3])
y = np.array([1,3,2,5])
f_interp = interp1d(x,y)
print('Interpolated at 1.5:', f_interp(1.5))
```

## Signal Processing & FFT

```python
t = np.linspace(0,1,500)
sig = np.sin(2*np.pi*5*t) + 0.5*np.random.randn(500)
fft_vals = np.fft.fft(sig)
freq = np.fft.fftfreq(len(t), t[1]-t[0])
plt.plot(freq[:250], np.abs(fft_vals[:250]))
plt.title('FFT Spectrum')
plt.show()
```

# Mini Project 3 — Gradient Descent Implementation

```python
def gradient_descent(lr=0.1, epochs=50):    x = 5    history = []    for _ in range(epochs):        grad = 2*x        x -= lr*grad        history.append(x)    return historyhistory = gradient_descent()plt.plot(history)plt.title('Gradient Descent Convergence')plt.show()
```
