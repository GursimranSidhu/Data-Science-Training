# OOPS : Object Oriented Programming

OOP organizes software design around data, or objects, rather than functions and logic. An object can be defined as a data field that has unique attributes and behavior.

comic-boom-oops-icon-simple-260nw-776058952.jpg


Object-oriented programming (OOP) is a programming paradigm that organizes code into objects, which are instances of classes. In Python, OOP allows you to structure your code around objects that have both data (attributes) and behavior (methods). It provides a way to model real-world entities and systems in a more intuitive and organized manner.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name}.")

    def celebrate_birthday(self):
        self.age += 1
        print(f"It's my birthday! I am now {self.age} years old.")


# Creating objects (instances) of the Person class
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)

# Accessing attributes
print(person1.name)  # Output: Alice
print(person2.age)   # Output: 30

# Invoking methods
person1.greet()            # Output: Hello, my name is Alice.
person2.celebrate_birthday()  # Output: It's my birthday! I am now 31 years old.
````

````python
age = int(6)
````

In this example, the `Person` class has two attributes: `name` and `age`. These attributes represent the state of the object and hold specific data for each instance of the class.

The class also has two methods: `greet()` and `celebrate_birthday()`. These methods define the behavior of the `Person` objects. The `greet()` method prints a personalized greeting message using the `name` attribute of the object. The `celebrate_birthday()` method increments the `age` attribute by 1 and prints a birthday message with the updated age.

By creating instances of the `Person` class (`person1` and `person2`), we can access their attributes (`name` and `age`) using dot notation (`object.attribute`). We can also invoke their methods using dot notation (`object.method()`), which performs actions or modifies the object's state.

In this example, `person1` has a name of "Alice" and an age of 25, while `person2` has a name of "Bob" and an age of 30. The objects can greet and celebrate their birthdays independently, showing different behavior based on their specific attributes.

**classes as numeric type**: the concept of classes as numeric types refers to the ability to create custom classes that behave like built-in numeric types (such as integers or floats). This is achieved by implementing special methods or magic methods that define the behavior of the class when used in numeric operations. By defining these special methods, you can customize how instances of your class behave when they participate in arithmetic operations, comparisons, and other numeric operations. This allows you to create custom numeric types that exhibit behavior specific to your application or domain.

````python
class Fraction:
    def __init__(self, numerator, denominator):
        self.numerator = numerator
        self.denominator = denominator

    def __add__(self, other):
        # Check if other is also a Fraction
        if isinstance(other, Fraction):
            # Calculate the common denominator
            common_denominator = self.denominator * other.denominator

            # Calculate the numerators based on the common denominator
            new_numerator = (self.numerator * other.denominator) + (other.numerator * self.denominator)

            # Create and return a new Fraction object
            return Fraction(new_numerator, common_denominator)
        else:
            raise ValueError("Unsupported operand type.")

    def __sub__(self, other):
        # Check if other is also a Fraction
        if isinstance(other, Fraction):
            # Calculate the common denominator
            common_denominator = self.denominator * other.denominator

            # Calculate the numerators based on the common denominator
            new_numerator = (self.numerator * other.denominator) - (other.numerator * self.denominator)

            # Create and return a new Fraction object
            return Fraction(new_numerator, common_denominator)
        else:
            raise ValueError("Unsupported operand type.")

    def __mul__(self, other):
        # Check if other is also a Fraction
        if isinstance(other, Fraction):
            # Multiply the numerators and denominators
            new_numerator = self.numerator * other.numerator
            new_denominator = self.denominator * other.denominator

            # Create and return a new Fraction object
            return Fraction(new_numerator, new_denominator)
        else:
            raise ValueError("Unsupported operand type.")

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"


# Creating instances of the Fraction class
fraction1 = Fraction(1, 2)
fraction2 = Fraction(4, 3)

# Performing arithmetic operations on instances
result1 = fraction1 + fraction2
result2 = fraction1 - fraction2
result3 = fraction1 * fraction2

# Displaying the results
print(result1)
print(result2)
print(result3)  # Output: 5/4
````

**Instance methods** take in an object as an arguemnt and have access to both class and object attributes

````python
class Rectangle:

    minimum_sides = 3

    def __init__(self, breadth, height):
        self.breadth = breadth
        self.height = height

    # this is an instance method
    def area(self):
        return self.breadth * self.height

    # getter and setter
    # we will explore in detail about getters and setters in encapsulation,
    # and why we need them
    def get_height(self):
        return self.height

    def set_height(self, n):
        self.height = n

    def get_breadth(self):
        return self.breadth

    def set_breadth(self, n):
        self.breadth = n

    # instance methods can also get and set class variables using self.__class__
    # to access class attributes that's why they are considered the most powerful
    def get_minimum_sides(self):
        return self.__class__.minimum_sides

    def set_minimum_sides(self, n):
        self.__class__.minimum_sides = n

r1 = Rectangle(4, 5)
print(f"The class Rectangle must have a minimum of {Rectangle.minimum_sides}")
print(r1.area())
print(r1.get_height())
print(r1.get_breadth())
print(r1.get_minimum_sides())
````

````python
r1.set_height(10)
r1.set_breadth(20)
r1.set_minimum_sides(4)

print(r1.area())
print(r1.get_height())
print(r1.get_breadth())
print(r1.get_minimum_sides())
````

**class methods**: takes in a class as an attribute and has access to all class attributes

````python
class Rectangle:

    minimum_sides = 3

    def __init__(self, breadth, height):
        self.breadth = breadth
        self.height = height

    # class methods are decorated with the classmethod decorator,
    # we'll explore decorators in depth in the decorators section
    @classmethod
    def get_minimum_sides(cls):
        return cls.minimum_sides

    @classmethod
    def set_minimum_sides(cls, n):
        cls.minimum_sides = n

r1 = Rectangle(4, 5)
print(r1.get_minimum_sides())
r1.set_minimum_sides(5)
print(r1.get_minimum_sides())
````

**static methods**: these methods don't take neither objects not classes as arguments. Thus, their scope of attributes is strongly bounded. Attributes and objects aren't available to them and they can't change the state of classes and objects. they can only operate on the arguments that were passed to them.  The main advantage of static methods that they **don't depend on classes and objects**. In addition, they can be called directly through a class without any objects creation. Calling these methods through objects is possible as well. We can define static methods with the help of a special decorator @staticmethod.

````python
import random
class Rectangle:

    def __init__(self, breadth, height):
        self.breadth = breadth
        self.height = height

    @staticmethod
    def get_info():
        return 'this is a Rectangle class'

    @staticmethod
    def get_random_number(start=0, end=100):
        '''
        returns a number between 0 and 100 unless specified
        '''
        return random.randint(start, end)

print(Rectangle.get_info())
print(Rectangle.get_random_number())
print(Rectangle.get_random_number(100, 1000))
````

The class interface serves as a contract between the class and its users, establishing the expected behavior and functionality of the class. It defines the operations that can be performed on instances of the class and the information that can be accessed or modified.

In Python, class interfaces are not explicitly defined or enforced by language constructs. Instead, they are typically documented through docstrings, comments, or class documentation. Python follows a principle known as "we are all consenting adults," trusting that developers will adhere to the documented class interface when using a class. Additionally, naming conventions (such as using underscores for internal attributes) can help indicate which attributes or methods are intended to be private or internal to the class.

````python
class Book:
    def __init__(self, title, author):
        # Initializing the attributes for the Book class
        self.title = title
        self.author = author
        self.is_checked_out = False  # Book is not checked out by default

    def __str__(self):
        # String representation of the book for easy printing
        return f'"{self.title}" by {self.author}'


class Library:
    def __init__(self):
        # Initialize an empty list to hold the books in the library
        self.books = []

    def add_book(self, book):
        # Add a book to the library's collection
        self.books.append(book)
        print(f'Added {book} to the library.')

    def remove_book(self, title):
        # Remove a book from the library by title
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                print(f'Removed {book} from the library.')
                return
        print(f'Book titled "{title}" not found in the library.')

    def check_out(self, title):
        # Check out a book by title
        for book in self.books:
            if book.title == title:
                if not book.is_checked_out:
                    book.is_checked_out = True  # Mark the book as checked out
                    print(f'Checked out {book}.')
                else:
                    print(f'Sorry, {book} is already checked out.')
                return
        print(f'Book titled "{title}" not found in the library.')

    def return_book(self, title):
        # Return a checked-out book by title
        for book in self.books:
            if book.title == title:
                if book.is_checked_out:
                    book.is_checked_out = False  # Mark the book as available
                    print(f'Returned {book}.')
                else:
                    print(f'The book {book} was not checked out.')
                return
        print(f'Book titled "{title}" not found in the library.')


# Example usage:
library = Library()  # Create a new library
book1 = Book("1984", "George Orwell")  # Create a new book
book2 = Book("To Kill a Mockingbird", "Harper Lee")  # Create another book


````

Constructor in inheritance: we can call parental methods in child methods and then process their results

````python
class Person:

    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def greet(self):
        print(f'Hello world! My name is {self.name}')


class Student(Person):

    def __init__(self, name, age, gender, standard):
        # this will call the parent class and pass the same arguments there
        # Person.__init__(self,name, age, gender)

        super().__init__( name, age, gender)

        # we can add extra fields than the parent class
        self.standard = standard


student1 = Student('John', 16, 'Male', '12th')
# although the child class didn't had the greet method but the parent class
# that it inherited from had, that's why there's no error
student1.greet()
````

**with the super method:** super does not refer to the parent class. It refers to the next class in the mro list, which depends on the actual class being instantiated. So when calling super().__init__, the actual method called is undetermined from the calling frame. so to instantiate the next parent constructor, another super will be called but this time it will take argument of the current parent constructor and the self argument

````python
# some parent classes
class Person:

    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def greet(self):
        print(f'Hello world! My name is {self.name}')


class Company:

    def __init__(self, company_name='Unknown'):
        self.company_name = company_name

    def description(self):
        print(f'I am an employee of {self.company_name}')


class Employee(Person,Company):

    def __init__(self, name, age, gender, company_name):
        super().__init__(name,age, gender)
        self.company_name = company_name
        self.employee_id = self._generate_employee_id()
        self.salary = self._calculate_salary()
        print(f"I am an employee with ID {self.employee_id} and a salary of {self.salary}.")

    def _generate_employee_id(self):
        # Add your logic to generate an employee ID
        return "EMP001"

    def _calculate_salary(self):
        # Add your logic to calculate the employee's salary
        return 5000

emp1 = Employee('John', 30, 'Male', 'Sabudh Foundation')
emp1.description()
emp1.greet()
````

### getters and setters

Getters and setters are methods used to access and modify the private
attributes of a class. They provide a controlled way to get and set the values of these attributes, which helps to enforce constraints or validation rules.


* Getter: a method that allows us to access an attribute in a given class
* Setter: a method that allows us to set or mutate the value of an attribute in a class

Implementing the getter and setter pattern requires:

1. Making your attributes non-public
2. Writing getter and setter methods for each attribute


````python
# a different example
class Employee:

    __company_name = 'Sabudh Foundation'
    __counter = 0 # a counter to keep a record of number of employees

    def __init__(self, name, age, employee_id):
        self.name = name
        self._age = age
        self._employee_id = employee_id

        self.__class__.__counter += 1

    @property
    # The @property decorator allows you to define a method that can be accessed like an attribute, meaning you can call it without parentheses.
    # This makes the code cleaner and more intuitive.
    def counter(self):
      # You first define a property using the @property decorator, which allows you to create a getter method.
      # This getter retrieves the current value of __counter, which is a class attribute that keeps track of the number of Employee instances.
        return self.__class__.__counter

    @counter.setter
    # After defining the getter, you can define a setter method for the same property using the @<property_name>.setter decorator.
    #  This method allows you to set a new value for the counter.
    def counter(self, n):
        self.__class__.__counter = n

    @property
    def company_name(self):
        return self.__class__.__company_name


emp1 = Employee('Jane Doe', 30, 'EMP01')
emp2 = Employee('John Doe', 30, 'EMP02')
````

# Employee Salary Management System

## Problem Statement

You are designing a system to manage different types of employees in a company. Each employee has a name and an ID, but the way their salary is calculated varies depending on their employment type. The company has three types of employees:

1. **Full-time Employee**: Receives a fixed monthly salary.
2. **Part-time Employee**: Receives an hourly wage multiplied by the number of hours worked.
3. **Contractor**: Receives a fixed fee per project.

## Requirements

### 1. Base Class

Create a base class named `Employee` with the following attributes and methods:

- **Attributes**:
  - `name` (string): The name of the employee.
  - `id` (string): The unique identifier for the employee.

- **Methods**:
  - `get_salary()` (abstract method): This method should be defined in subclasses to calculate and return the employee's salary.

### 2. Derived Classes

Create the following derived classes from `Employee`:

#### a. FullTimeEmployee

- **Attributes**:
  - `salary` (float): The fixed monthly salary.

- **Methods**:
  - Override `get_salary()` to return the fixed salary.

#### b. PartTimeEmployee

- **Attributes**:
  - `hourly_wage` (float): The rate per hour.
  - `hours_worked` (float): The total number of hours worked in a month.

- **Methods**:
  - Override `get_salary()` to return `hourly_wage * hours_worked`.

#### c. Contractor

- **Attributes**:
  - `fee_per_project` (float): The fixed fee received per project.
  - `projects_completed` (int): The number of projects completed.

- **Methods**:
  - Override `get_salary()` to return `fee_per_project * projects_completed`.

### 3.

Create a function `print_salary(employee)` that:

- **Parameters**:
  - `employee` (Employee): An instance of the `Employee` class or its subclasses.

- **Functionality**:
  - This function should work with any derived class of `Employee` and correctly calculate and display the salary.

### Example Usage

```python
full_time = FullTimeEmployee("Alice", "F001", 5000)
part_time = PartTimeEmployee("Bob", "P002", 20, 80)
contractor = Contractor("Charlie", "C003", 1500, 5)

print_salary(full_time)     # Output: Alice (ID: F001) earns $5000
print_salary(part_time)     # Output: Bob (ID: P002) earns $1600
print_salary(contractor)   # Output: Charlie (ID: C003) earns $7500


````python
from abc import ABC, abstractmethod

# Base Class
class Employee(ABC):
    def __init__(self, name, id):
        self.name = name
        self.id = id

    @abstractmethod
    def get_salary(self):
        pass

# Derived Class: FullTimeEmployee
class FullTimeEmployee(Employee):
    def __init__(self, name, id, salary):
        super().__init__(name, id)
        self.salary = salary

    def get_salary(self):
        return self.salary

# Derived Class: PartTimeEmployee
class PartTimeEmployee(Employee):
    def __init__(self, name, id, hourly_wage, hours_worked):
        super().__init__(name, id)
        self.hourly_wage = hourly_wage
        self.hours_worked = hours_worked

    def get_salary(self):
        return self.hourly_wage * self.hours_worked

# Derived Class: Contractor
class Contractor(Employee):
    def __init__(self, name, id, fee_per_project, projects_completed):
        super().__init__(name, id)
        self.fee_per_project = fee_per_project
        self.projects_completed = projects_completed

    def get_salary(self):
        return self.fee_per_project * self.projects_completed

# Function to print salary
def print_salary(employee):
    print(f"{employee.name} (ID: {employee.id}) earns ${employee.get_salary()}")

# Example Usage
full_time = FullTimeEmployee("Alice", "F001", 5000)
part_time = PartTimeEmployee("Bob", "P002", 20, 80)
contractor = Contractor("Charlie", "C003", 1500, 5)

print_salary(full_time)     # Output: Alice (ID: F001) earns $5000
print_salary(part_time)     # Output: Bob (ID: P002) earns $1600
print_salary(contractor)   # Output: Charlie (ID: C003) earns $7500
````

Access Specifiers



Public → Free to use anywhere (owner, deposit, get_balance).

Protected → Accessible but meant for internal/child classes (_account_type, _get_account_type).

Private → Truly hidden, only class itself should use it (__balance, __update_balance).

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner        # Public attribute
        self._account_type = "Savings"  # Protected attribute (internal use)
        self.__balance = balance  # Private attribute (hidden)

    # Public method
    def deposit(self, amount):
        if amount > 0:
            self.__update_balance(amount)   # calls private method internally
            print(f"Deposited {amount}, new balance: {self.__balance}")
        else:
            print("Invalid deposit amount")

    # Protected method
    def _get_account_type(self):
        return self._account_type

    # Private method
    def __update_balance(self, amount):
        # Only accessible inside the class
        self.__balance += amount

    # Public getter for balance
    def get_balance(self):
        return self.__balance


# Usage
account = BankAccount("Alice", 1000)

# Public attribute and method
print(account.owner)        # ✅ "Alice"
account.deposit(500)        # ✅ Deposits money
print(account.get_balance()) # ✅ 1500

# Protected attribute and method (accessible but not recommended)
print(account._account_type)         # ⚠️ Works: "Savings"
print(account._get_account_type())   # ⚠️ Works but meant for internal use



````

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner                 # Public
        self._account_type = "Savings"     # Protected
        self.__balance = balance           # Private

    # Public method
    def deposit(self, amount):
        if amount > 0:
            self.__update_balance(amount)  # calls private method internally
            print(f"Deposited {amount}, new balance: {self.__balance}")
        else:
            print("Invalid deposit amount")

    # Protected method
    def _get_account_type(self):
        return self._account_type

    # Private method
    def __update_balance(self, amount):
        self.__balance += amount

    # Public getter
    def get_balance(self):
        return self.__balance


# Subclass
class PremiumAccount(BankAccount):
    def __init__(self, owner, balance, rewards_points=0):
        super().__init__(owner, balance)
        self.rewards_points = rewards_points  # Public

    def show_details(self):
        # ✅ Can access public attribute
        print(f"Owner: {self.owner}")

        # ✅ Can access protected attribute
        print(f"Account Type: {self._account_type}")

        # ❌ Cannot directly access private attribute
        # print(self.__balance)   # AttributeError

        # ✅ Can use public method of parent to get balance
        print(f"Balance: {self.get_balance()}")

        # ✅ Can call protected method
        print(f"Type via method: {self._get_account_type()}")

        # ❌ Cannot call private method of parent
        # self.__update_balance(500)  # AttributeError


# Usage
acc = PremiumAccount("Bob", 2000)

acc.show_details()
acc.deposit(1000)  # Uses parent class public method
print("Rewards:", acc.rewards_points)
````

## Data Abstraction in Python

Abstraction is the concept of hiding implementation details and exposing only the essential features to the user.

It focuses on what an object does rather than how it does it.

In Python, abstraction is commonly implemented using abstract classes and abstract methods (via the abc module).

An abstract class cannot be instantiated; it only provides a blueprint for its subclasses.

Subclasses must implement the abstract methods.

from abc import ABC, abstractmethod
* **abc module:**
This is a built-in Python module that provides the infrastructure for defining Abstract Base Classes.
* **ABC:**
This stands for "Abstract Base Class." When you define a class that inherits from ABC, you are explicitly declaring it as an abstract class. Abstract classes cannot be instantiated directly; they serve as blueprints for other classes.
* **abstractmethod:**
This is a decorator used to mark methods within an abstract class as "abstract." An abstract method is declared in the abstract class but does not have an implementation there. Any concrete (non-abstract) class that inherits from this abstract class must provide an implementation for all abstract methods defined in the parent ABC. If a subclass fails to implement an abstract method, Python will raise a TypeError when you attempt to instantiate that subclass.

````python
from abc import ABC, abstractmethod

# Abstract class
class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass   # abstract method (must be implemented by child classes)

    def display():
      print("hello")


# Concrete subclass 1
class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using Credit Card.")


# Concrete subclass 2
class UpiPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using UPI.")


# Usage
# payment = Payment()  # ❌ TypeError: Can't instantiate abstract class

payment1 = CreditCardPayment()
payment1.pay(1000)  # ✅ "Paid 1000 using Credit Card."

payment2 = UpiPayment()
payment2.pay(500)   # ✅ "Paid 500 using UPI."
````

| Feature            | **Abstraction**                                                                                            | **Encapsulation**                                                                                                              |
| ------------------ | ---------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------ |
| **Meaning**        | Hiding **implementation details** and showing only the essential features.                                 | Hiding **data/variables** of a class and restricting direct access to them.                                                    |
| **Focus**          | Focuses on **what** an object does, not **how** it does it.                                                | Focuses on **how to protect the object’s state** (data).                                                                       |
| **Implementation** | Implemented using **abstract classes** and **interfaces** (`abc` module in Python).                        | Implemented using **access specifiers** (public, protected `_`, private `__` in Python).                                       |
| **Example**        | Abstract class `Shape` defines `area()`, subclasses (`Circle`, `Rectangle`) implement it differently.      | A `BankAccount` class hides its `__balance` and exposes `deposit()` / `get_balance()` methods.                                 |
| **Analogy**        | **Car driver** only knows how to use *accelerator, brake, steering* (what it does), not the engine design. | **Car engine** is locked under the hood — internal details are hidden, you can only use the interface provided (key, buttons). |
| **Goal**           | To achieve **design-level hiding** (hide implementation).                                                  | To achieve **data-level hiding** (hide internal state).                                                                        |
