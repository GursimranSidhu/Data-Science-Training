# RESTful API Concepts

An API (Application Programming Interface) allows different software systems to communicate with each other. In web development, APIs allow a client application (browser, mobile app, or script) to request data or services from a server.

# Working Princepal
Client  →  API Request  →  Server


Client  ←  API Response ←  Server

# RESTful Api


A RESTful API is an API that follows the architectural principles of REST (Representational State Transfer).

REST

RESTful APIs use standard HTTP methods to perform operations on resources such as users, products, or orders.

Example API endpoint:

https://jsonplaceholder.typicode.com/users

Here, users is a resource that can be accessed or modified.

# Key Characteristics of RESTful APIs


1 Client–Server Architecture

The client and server operate independently.

Client: sends requests

Server: processes requests and returns responses

Example:

Mobile App → API Server → Database


2 Stateless Communication

In REST APIs, each request must contain all the information needed to process it.

The server does not store client session information between requests.

Example:

GET /users/10
Authorization: Bearer token

Each request includes authentication information.



3 Resource-Based Design

REST APIs treat everything as a resource.

Examples:

Resource	Endpoint
Users	/users
Products	/products
Orders	/orders

Example request:

GET /users/5

This retrieves the user with ID 5.


4 Uniform Interface

REST APIs use a standard set of HTTP methods to interact with resources.

Method	Purpose
GET	Retrieve data
POST	Create new resource
PUT	Update resource
PATCH	Partial update
DELETE	Remove resource

Example:

GET /products
POST /products
PUT /products/10
DELETE /products/10

Web scraping is widely used for:

Data analysis

Price comparison

Automation tasks

Most websites display content using HTML, CSS, and JavaScript. Web scraping tools help us download the page and extract useful information from its structure.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Start browser
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# Open Flipkart search page
driver.get("https://www.flipkart.com/search?q=iphone")

time.sleep(3)

# Close login popup
try:
    close_btn = driver.find_element(By.XPATH, "//button[contains(text(),'✕')]")
    close_btn.click()
except:
    pass

time.sleep(3)

# Find iPhone titlesresponse
titles = driver.find_elements(By.CLASS_NAME, "_4rR01T")

# Find prices
prices = driver.find_elements(By.CLASS_NAME, "_30jeq3")

# Print results
for title, price in zip(titles, prices):
    print(title.text, ":", price.text)

driver.quit()
```

```python
import scrapy
from scrapy.crawler import CrawlerRunner
from twisted.internet import reactor, defer
import nest_asyncio

nest_asyncio.apply()

class ExampleSpider(scrapy.Spider):
    name = "example"
    start_urls = ["https://www.flipkart.com/search?q=iphone"]

    def parse(self, response):
        title = response.css("title::text").get()
        yield {"title": title}

runner = CrawlerRunner(settings={
    "REQUEST_FINGERPRINTER_IMPLEMENTATION": "2.7",
    "USER_AGENT": "Mozilla/5.0"
})

@defer.inlineCallbacks
def crawl():
    yield runner.crawl(ExampleSpider)
    reactor.stop()

crawl()
reactor.run()
```
