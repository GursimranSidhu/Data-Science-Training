# Type Conversion

Type conversion in Python refers to changing a value 
from one data type to another. This can happen automatically by
Python or manually by the programmer using built-in functions. 
Your code demonstrates both forms of type conversion.

Key Points

Implicit type conversion happens automatically when Python converts one data type to another during an operation.

Explicit type conversion (type casting) is done manually using functions like int(), float(), str(), list(), and set().

Converting between sequence types (string → list → set) changes how the data is stored and accessed.

```python
# Type Conversion 
calc =  10
print(type(calc),calc)
calc = calc+1.25
print(type(calc),calc)
calc=  str(calc)
print(type(calc),calc)
calc =list(calc)
print(type(calc),calc)
calc=tuple(calc)
print(type(calc),calc)
calc= set(calc)
print(type(calc),calc)
# order here is random there {'.', '2', '5', '1'}
# {'1', '5', '2', '.'}
# {'2', '.', '1', '5'} Hash randomization
```

# Indentation

Indentation in Python refers to the use of spaces (or tabs) at the beginning of a line to define blocks of code. Unlike many other programming languages, Python uses indentation instead of braces {} to control the structure and flow of a program.

Points

Indentation shows which statements belong to the same block of code.

It is mandatory in Python; incorrect indentation causes errors.

The standard practice is to use 4 spaces for each level of indentation.

```python
# Control Structures
# python uses indentation instead of {} to define code blocks
# Example
y = True
if y:
    print('I used 4 spaces')
```

# While Loops
Overview

A while loop repeatedly executes a block of code as long as a condition is true. It is useful when the number of iterations is not known in advance.

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

# Counting with While Loops
Overview

Counting is commonly done by updating a counter variable inside the loop.

Example

```python
num = 0

while num < 10:
    print(num)
    num += 2
```

```python
num
```

# For Loops
Overview

A for loop is used to 
iterate over a sequence such as a list,
string, or range. It is best when the number of iterations is known.

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

# Counting Using range() Function
Overview

The range() function generates a sequence of numbers and is commonly used with for loops.

Example

```python
count =0
for i in range(10,2,-3):
    count+=1
    print(count,i)
```

# While vs. For Loops
Overview

Use while when the condition depends on dynamic input.

Use for when iterating over a sequence or fixed range.

```python
# while loop
x = 3
while x > 0:
    print(x)
    x -= 1

# for loop
for x in range(3, 0, -1):
    print(x)
```

# Nested Loops
Overview

A nested loop is a loop inside another loop. Commonly used in patterns and tables.

```python
students= 38
for i in range(5):
    while students > 35:
        students-=1
```

```python
students
```

# break Statement
Overview

break is used to exit the loop immediately when a condition is met.

```python
students= 32
for i in range(5):
    students+=1
    if students > 35:
        break;
```

```python
students
```
