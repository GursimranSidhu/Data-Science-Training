## 2. Returning Values from Functions

What is return in Detail?

The return statement:

Sends a value back to where the function was called

Stops the function immediately

Can return any data type (number, string, list, etc.)

## 6. Function Arguments

```python
def summations(a,b):
```

```python

def introduce(name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce("John", 25)
introduce(age=30, name="Emma")

def power(base, exponent=2):
    return base ** exponent

print(power(5))
print(power(5, 3))

def sum_all(*numbers):
    return sum(numbers)

print(sum_all(1, 2, 3, 4))
```

## 8. Anonymous Functions (Lambda)

```python

square_lambda = lambda x: x * x
print(square_lambda(5))

add_lambda = lambda a, b: a + b
print(add_lambda(3, 7))
```

## 9. map(), filter(), and reduce()

```python

numbers = [1, 2, 3, 4, 5]

squared = list(map(lambda x: x**2, numbers))
print("Squared:", squared)

evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Evens:", evens)

from functools import reduce

sum_total = reduce(lambda a, b: a + b, numbers)
print("Sum using reduce:", sum_total)
```
