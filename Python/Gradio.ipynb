```python
import gradio as gr
import numpy as np
import numpy as np

def grayscale(image):
    gray = np.mean(image, axis=2)

    # Convert to uint8 (0–255)
    gray = gray.astype(np.uint8)

    return gray

app = gr.Interface(
    fn=grayscale,
    inputs="image",
    outputs="image"
)

app.launch()
```

```python
import gradio as gr

def process(name, age):
    return f"{name} is {age} years old"

demo = gr.Interface(
    fn=process,
    inputs=[gr.Textbox(label="Name"), gr.Slider(1, 100)],
    outputs="text"
)

demo.launch()
```

```python
import gradio as gr

with gr.Blocks() as demo:
    with gr.Row():  #check with Column Tab 
        name = gr.Textbox(label="Name")
        age = gr.Number(label="Age")

    output = gr.Textbox()

    btn = gr.Button("Submit")

    btn.click(fn=lambda n, a: f"{n} is {a} years old",
              inputs=[name, age],
              outputs=output)

demo.launch()
```

```python
import gradio as gr

def square(x):
    return x * x

with gr.Blocks() as demo:
    num = gr.Number()
    out = gr.Number()
    btn = gr.Button("Square")

    btn.click(fn=square, inputs=num, outputs=out)

demo.launch()
```

```python
# change even 
import gradio as gr

def update_text(x):
    return f"You typed: {x}"

with gr.Blocks() as demo:
    text = gr.Textbox()
    output = gr.Textbox()

    text.change(fn=update_text, inputs=text, outputs=output)

demo.launch()
```

```python
# submit 
import gradio as gr

def respond(text):
    return text.upper()

with gr.Blocks() as demo:
    textbox = gr.Textbox()
    output = gr.Textbox()

    textbox.submit(fn=respond, inputs=textbox, outputs=output)

demo.launch()
```

```python
# more than one functions 

import gradio as gr

def add(x, y):
    return x + y

def multiply(x, y):
    return x * y
def substract(x,y):
    return x-y
def divison(x,y):
    return x/y

with gr.Blocks() as demo:
    x = gr.Number()
    y = gr.Number()

    sum_out = gr.Number()
    prod_out = gr.Number()
    reduced_out = gr.Number()
    divison_out = gr.Number()

    btn = gr.Button("Calculate")

    btn.click(fn=add, inputs=[x, y], outputs=sum_out)
    btn.click(fn=multiply, inputs=[x, y], outputs=prod_out)
    btn.click(fn=substract, inputs=[x, y], outputs=reduced_out)
    btn.click(fn=divison, inputs=[x, y], outputs=divison_out)

demo.launch()
```

```python
import gradio as gr

def calculate(a, b, operation):
    if operation == "Add":
        return a + b
    elif operation == "Multiply":
        return a * b
    elif operation == "Subtract":
        return a - b
    elif operation == "Divide":
        return a / b

with gr.Blocks() as demo:
    gr.Markdown("## Simple Calculator")

    a = gr.Number(label="Number 1")
    b = gr.Number(label="Number 2")

    op = gr.Dropdown(["Add", "Multiply", "Subtract", "Divide"])

    result = gr.Number()

    btn = gr.Button("Calculate")

    btn.click(fn=calculate,
              inputs=[a, b, op],
              outputs=result)

demo.launch(share=True)
```

```python
import gradio as gr
from openai import OpenAI

client = OpenAI(api_key = "your api key")

def chat_with_ai(message, history):
    messages = []

    # Convert history into OpenAI format
    for user, bot in history:
        messages.append({"role": "user", "content": user})
        messages.append({"role": "assistant", "content": bot})

    # Add current message
    messages.append({"role": "user", "content": message})

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )

    reply = response.choices[0].message.content
    return reply

# Gradio UI
app = gr.ChatInterface(
    fn=chat_with_ai,
    title="My ChatGPT Clone 🤖",
    description="Ask me anything!"
)

app.launch()
```
