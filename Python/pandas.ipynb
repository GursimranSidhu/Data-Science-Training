### What is Pandas

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool,
built on top of the Python programming language.

https://pandas.pydata.org/about/index.html

#### Pandas is an open-source Python library designed for data manipulation and analysis. It provides easy-to-use data structures (mainly Series for one-dimensional data and DataFrame for two-dimensional, tabular data) to work efficiently with structured data, such as tables from spreadsheets, databases, or CSV files.

### What to Use Pandas For
* Loading Data: Import data from files (CSV, Excel, SQL, JSON, etc.) into a DataFrame.

* Data Cleaning: Handle missing data, remove duplicates, and fix errors in your dataset.

* Transforming Data: Filter, sort, group, merge, or reshape data to suit the needs of your analysis.

* Analyzing Data: Compute statistics (mean, sum, correlations), aggregate data, and generate summary tables.

* Visualization: Integrate with plotting libraries (like Matplotlib) to create charts and graphs easily.

* Exporting Data: Save cleaned and analyzed results back to various formats.

### Why Use Pandas?
* Simplicity: Makes it easy to manipulate, clean, and analyze data using Python.

* Power: Supports advanced data aggregation, grouping, and statistical analysis.

* Integration: Works seamlessly with other scientific libraries like NumPy, Matplotlib, and Scikit-learn.

* Performance: Efficient handling of large datasets for real-world data science tasks.

* Versatility: Used across industries (finance, healthcare, marketing, and more) for analytics and decision making.

In summary, use Pandas when you need to work with and analyze structured data efficiently in Python—from cleaning raw data to building insightful summaries or preparing datasets for machine learning models

## DataFrame

DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.

pandas_df_first_look.png

DataFrame vs Series: What's the difference ?

Screenshot 2021-01-20 112630.png


Series is a list that can store values of different datatypes whereas a DataFrame is a collection of series. Each column in a DataFrame is a Series in itself.

```python
import pandas as pd
```

```python
# Method-1 using list
list_df = pd.DataFrame(data=[[1, 2, 3, 4], [5, 6, 7, 8]], columns=["one", "two", "three", "four"])
list_df.head()
```

```python
import pandas as pd

# Example DataFrame with date as string
data = {
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "join_date": ["15/01/2025", "22/03/2025", "19/08/2025"]  # DD/MM/YYYY format
}

df = pd.DataFrame(data)

print("Before parsing:")
print(df.dtypes)
print(df)

# Use lambda + apply to parse date strings
df["join_date"] = df["join_date"].apply(lambda x: pd.to_datetime(x, dayfirst=True))

print("\nAfter parsing:")
print(df.dtypes)
print(df)
```
