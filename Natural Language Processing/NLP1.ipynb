```python
import nltk
import string

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
text="Natural Language Processing (NLP) enables computers to understand human languagae."
text
```

**1. Lower Case Conversion**

```python
text_lower=text.lower()
text_lower
```

**2. Tokenization**

```python
tokens=word_tokenize(text_lower)
print(tokens)
```

**3. Punctuation Removal**

```python
tokens_no_punct= [word for word in tokens if word not in string.punctuation]
print(tokens_no_punct)
```

**4. Stopword Removal**

```python
stop_words=set(stopwords.words('english'))
tokens_no_stopwords=[word for word in tokens_no_punct if word not in stop_words]
print(tokens_no_stopwords)
```

**5. Stemming**

```python
stemmer=PorterStemmer()
stemmed_tokens= [stemmer.stem(word) for word in tokens_no_stopwords]
print(stemmed_tokens)
```

**6. Lemmatization**

```python
lemmatizer=WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(word) for word in tokens_no_stopwords]
print(lemmatized_tokens)
```

```python
lemmatizer.lemmatize("running",pos='n')
```

```python
lemmatizer.lemmatize("running",pos='v')
```

**FINAL PROCESSED TEXT**

```python
processed_text=" ".join(lemmatized_tokens)
processed_text
```

**7. Bag Of Words**



*   Bow_matrix shows how many times each vocabulary word appears in each document.
*   CountVectorizer performs basic preprocessing like lowercasing and regex based tokenization, but it do not handle linguistic normalization such as stemming or lemmatization. Those must be done explicitly.




```python
corpus= ["Natural Language Processing (NLP) enables computers to understand human languagae.",
"Language Processing is a key part of Artificial Intelligence.",
"Artificial Intellgence is emergence of modern world."]

vectorizer= CountVectorizer(lowercase=True, stop_words='english')
bow_matrix=vectorizer.fit_transform(corpus)

print("Vocbulary: ")
print(vectorizer.get_feature_names_out())

print("\nBag of Words Martrix: ")
print(bow_matrix.toarray())
```

**8. TF-IDF Representation**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus= ["Natural Language Processing (NLP) enables computers to understand human languagae.",
"Language Processing is a key part of Artificial Intelligence.",
"Artificial Intellgence is emergence of modern world."]

vectorizer=TfidfVectorizer()

tfidf_matrix=vectorizer.fit_transform(corpus)

# Vocbulary
feature_names=vectorizer.get_feature_names_out()

print("Vocabulary: ")
print(feature_names)

print("\nTF_IDF matrix: ")
print(tfidf_matrix.toarray())
```
